17  What is the difference between a class variable and an instance variable?
-In object-oriented programming, **class variables** and **instance variables** are both used to store data, but they have important differences in terms of their scope, behavior, and how they are accessed.

### 1. **Class Variables:**
   - **Definition:** Class variables are variables that are shared among all instances (objects) of a class. They are defined within the class but outside any methods.
   - **Scope:** They are accessible via both the class itself and instances of the class.
   - **Shared Among Instances:** All instances of the class refer to the same memory location for a class variable. If the class variable is modified via one instance, the change is reflected in all other instances.
   - **Access:** Can be accessed directly using the class name (e.g., `ClassName.variable`) or through any instance (e.g., `instance.variable`).
   - **Use Case:** Used for values that are the same for all instances of the class (e.g., a constant or a counter shared by all objects).

   **Example:**
   ```python
   class Car:
       wheels = 4  # class variable

   car1 = Car()
   car2 = Car()

   print(car1.wheels)  # 4
   print(car2.wheels)  # 4
   ```

   In this case, both `car1` and `car2` share the same class variable `wheels`.

### 2. **Instance Variables:**
   - **Definition:** Instance variables are variables that are specific to each instance (object) of the class. They are typically defined within the `__init__()` method (the constructor) and are prefixed with `self`.
   - **Scope:** They are accessible only through instances of the class.
   - **Not Shared Among Instances:** Each instance has its own copy of the instance variable, so changes made to one instance's variable do not affect others.
   - **Access:** Accessed via the instance of the class (e.g., `instance.variable`).
   - **Use Case:** Used to store data that is unique to each instance of the class (e.g., attributes that vary for each object).

   **Example:**
   ```python
   class Car:
       def __init__(self, color):
           self.color = color  # instance variable

   car1 = Car('red')
   car2 = Car('blue')

   print(car1.color)  # red
   print(car2.color)  # blue
   ```

   Here, `car1` has an instance variable `color` with the value `'red'`, while `car2` has an instance variable `color` with the value `'blue'`.

### Summary of Key Differences:

| Feature                 | **Class Variable**                                       | **Instance Variable**                                     |
|-------------------------|----------------------------------------------------------|-----------------------------------------------------------|
| **Defined**             | Defined directly inside the class, outside methods.      | Defined inside the `__init__()` method (or other methods). |
| **Shared or Unique**    | Shared among all instances of the class.                 | Unique to each instance of the class.                     |
| **Access**              | Can be accessed using the class name or through instances. | Accessed via instances of the class (using `self`).       |
| **Scope**               | Accessible by both the class and instances.              | Accessible only through instances of the class.           |

In short, **class variables** are shared across all instances, while **instance variables** hold data unique to each instance.

19  Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python
-In Python, the `__str__` and `__repr__` methods are special methods used to define how an object is represented as a string. Both methods serve different purposes:

### 1. `__str__` Method
The `__str__` method is used to define a user-friendly or informal string representation of an object. It is primarily intended to be read by humans. When you use the `print()` function or the `str()` function on an object, Python will call the `__str__` method (if it’s defined) to return a string that is easily understandable.

- **Purpose**: Provide a readable, easy-to-understand string representation of an object.
- **Usage**: Primarily for displaying the object to the user (e.g., printing the object).

#### Example:
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __str__(self):
        return f"Person: {self.name}, {self.age} years old"
        
person = Person("John", 30)
print(person)  # Output: Person: John, 30 years old
```

Here, when we call `print(person)`, Python uses the `__str__` method to display a user-friendly description of the `Person` object.

### 2. `__repr__` Method
The `__repr__` method is used to define a more formal or unambiguous string representation of an object. Its primary goal is to provide a string that can be used to recreate the object or at least give detailed information about it. The output from `__repr__` should ideally be something that, if passed to the `eval()` function, would return the original object (although this is not a strict requirement).

- **Purpose**: Provide a detailed and unambiguous representation of the object, usually for debugging or development purposes.
- **Usage**: Used by the `repr()` function, and is also the default string representation in interactive Python sessions.

#### Example:
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __repr__(self):
        return f"Person('{self.name}', {self.age})"
        
person = Person("John", 30)
print(repr(person))  # Output: Person('John', 30)
```

In this case, when you call `repr(person)`, Python uses the `__repr__` method to give a more detailed string representation of the `Person` object.

### Key Differences:
- `__str__`: Intended for human-readable output (e.g., in `print()` or `str()`).
- `__repr__`: Intended for a more detailed or unambiguous representation, typically useful for developers or debugging (e.g., in the interpreter or when using `repr()`).

### When Both Are Defined:
- If you define both methods in a class, `__str__` will be used by functions like `print()` and `str()`, while `__repr__` will be used in contexts that require a detailed representation of the object, such as the Python interactive shell.
- If you don't define `__str__`, Python will fall back on `__repr__`.

#### Example with Both:
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def __str__(self):
        return f"Person: {self.name}, {self.age} years old"
        
    def __repr__(self):
        return f"Person('{self.name}', {self.age})"
        
person = Person("John", 30)
print(person)       # Output: Person: John, 30 years old (calls __str__)
print(repr(person)) # Output: Person('John', 30) (calls __repr__)
```

In summary:
- Use `__str__` for a readable, user-friendly string representation.
- Use `__repr__` for a more detailed, unambiguous representation suitable for debugging or development.

21 What is the significance of the __del__ method in Python?
-The `__del__` method in Python is a special method, often referred to as a *destructor*. It is called when an object is about to be destroyed or deallocated. In other words, when an object goes out of scope and is garbage collected, Python automatically calls the `__del__` method if it's defined in the object's class.

### Purpose and Significance:
1. **Resource Cleanup**: The `__del__` method is commonly used to perform cleanup tasks, such as releasing external resources (e.g., closing files, network connections, or database connections) that the object may have acquired during its lifetime. Python's garbage collector automatically handles most memory management, but it does not automatically clean up non-memory resources, which is where `__del__` comes in.

2. **Finalization**: It allows developers to define custom behavior that should occur just before an object is destroyed, which can be crucial for applications that rely on cleanup actions when objects are no longer needed.

### Example:

```python
class FileHandler:
    def __init__(self, filename):
        self.filename = filename
        self.file = open(filename, 'w')
        print(f"File {self.filename} opened")

    def write(self, content):
        self.file.write(content)
    
    def __del__(self):
        self.file.close()  # Ensures file is closed when object is destroyed
        print(f"File {self.filename} closed")

# Example usage
handler = FileHandler('example.txt')
handler.write('Hello, world!')
# When 'handler' goes out of scope or program ends, __del__ is called
```

### Key Points to Keep in Mind:
1. **Unreliable with Circular References**: If an object is part of a reference cycle (e.g., object A references object B, and B references A), `__del__` may not be called automatically. Python's garbage collector may not collect such cycles immediately, and thus the cleanup process might be delayed or skipped. To handle this, you may need to use the `gc` module for manual garbage collection.

2. **Not Guaranteed to Be Called**: The `__del__` method is not guaranteed to be invoked immediately when an object is no longer needed. In certain cases (like when the program terminates), Python may not call `__del__` at all, or it may be called late, depending on the garbage collection timing.

3. **Use with Caution**: It’s often better to use explicit resource management techniques like context managers (`with` statements) rather than relying solely on `__del__`, as they offer a more predictable and explicit way to manage resources.

### Context Manager Alternative:
A more robust and predictable approach to resource cleanup in Python is to use the context manager pattern with the `with` statement, which automatically handles the entry and exit of a block of code.

```python
class FileHandler:
    def __init__(self, filename):
        self.filename = filename
        self.file = open(filename, 'w')

    def write(self, content):
        self.file.write(content)

    def close(self):
        self.file.close()

    def __enter__(self):
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        self.close()

# Using the context manager
with FileHandler('example.txt') as handler:
    handler.write('Hello, world!')
# File is automatically closed when the block is exited
```

### Conclusion:
The `__del__` method is useful for resource cleanup when an object is about to be destroyed, but it should be used with caution due to potential issues with garbage collection and object lifecycle. When possible, consider using context managers for more reliable and explicit resource management.



22 What is the difference between @staticmethod and @classmethod in Python?
-In Python, both `@staticmethod` and `@classmethod` are used to define methods that are bound to a class rather than an instance of the class. However, they have different purposes and behaviors:

### 1. **@staticmethod**:
- **No access to instance (`self`) or class (`cls`)**: A `staticmethod` does not take `self` or `cls` as its first parameter. It behaves like a regular function but belongs to the class's namespace.
- **Used for utility functions**: Typically, `staticmethod` is used for utility functions that do not need access to instance or class-specific data. These functions are logically related to the class but don't need to interact with the instance or class itself.
- **Called on class or instance**: You can call a `staticmethod` on the class itself or on an instance of the class.

**Example**:
```python
class MyClass:
    @staticmethod
    def my_static_method(x, y):
        return x + y

# Usage
result = MyClass.my_static_method(3, 4)  # Called on the class
print(result)  # Output: 7
```

### 2. **@classmethod**:
- **Access to the class (`cls`)**: A `classmethod` takes `cls` as its first parameter, which refers to the class itself, not an instance. This allows the method to access class-level attributes and methods.
- **Can modify class state**: Since the method has access to the class, it can modify class-level variables or call other class methods.
- **Called on class or instance**: Like `staticmethod`, `classmethod` can be called on both the class and an instance of the class.

**Example**:
```python
class MyClass:
    count = 0  # A class variable

    @classmethod
    def increment_count(cls):
        cls.count += 1

# Usage
MyClass.increment_count()  # Called on the class
print(MyClass.count)  # Output: 1
```

### Key Differences:

| Feature           | `@staticmethod`                           | `@classmethod`                             |
|-------------------|-------------------------------------------|--------------------------------------------|
| **First Parameter** | None (no `self` or `cls`)                 | Takes `cls` (the class itself)             |
| **Access**         | Cannot access instance or class variables | Can access and modify class variables      |
| **Use Case**       | Utility functions that don't modify class/instance state | Methods that need to access or modify class state |
| **Callability**    | Can be called on both the class or an instance | Can be called on both the class or an instance |

### When to use which:
- Use `@staticmethod` when you want a method that doesn't depend on the instance or class.
- Use `@classmethod` when you need to modify or access class-level variables or perform class-specific operations.



23 How does polymorphism work in Python with inheritance?
-Polymorphism in Python, especially when combined with inheritance, is a concept that allows objects of different classes to be treated as objects of a common base class. In Python, polymorphism can be achieved through method overriding, where a method in a subclass has the same name as one in the parent class, but provides a different implementation.

Here’s how polymorphism works in Python with inheritance:

### 1. **Inheritance**:
   - Polymorphism typically relies on inheritance. A child class inherits methods and properties from a parent class, and it can override or extend these methods.

### 2. **Method Overriding**:
   - Polymorphism in Python allows the subclass to provide its own version of a method that was originally defined in the parent class. This is called *method overriding*. When an instance of the subclass calls the method, the overridden version in the subclass is executed, not the one in the parent class.

### 3. **Dynamic Method Dispatch**:
   - Python uses dynamic method dispatch to decide which method to call. When a method is called on an object, Python first checks if the object’s class has the method. If it does, that method is called; otherwise, it searches up the class hierarchy.

Here’s an example to illustrate how polymorphism works in Python:

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

# Polymorphism in action
def make_animal_speak(animal: Animal):
    animal.speak()

# Create instances
dog = Dog()
cat = Cat()

# Call the same method on different types of objects
make_animal_speak(dog)  # Outputs: Dog barks
make_animal_speak(cat)  # Outputs: Cat meows
```

### Key Points:
1. **Base Class**: `Animal` has a method `speak()`, which is a general definition.
2. **Derived Classes**: `Dog` and `Cat` both inherit from `Animal` and override the `speak()` method to give their own implementation.
3. **Polymorphism in Action**: The function `make_animal_speak()` accepts any object of type `Animal`, and when passed either a `Dog` or `Cat` object, it calls the appropriate overridden method. Despite the function call being the same, the output depends on the type of object passed.

### Advantages of Polymorphism:
- **Flexibility**: You can write functions that work with objects of different classes without needing to know the specific type of the object.
- **Code Reusability**: You can use common interfaces (base classes) and reuse the code in different subclasses with specialized behaviors.

### In Conclusion:
Polymorphism with inheritance in Python allows for more generic and flexible code, as methods and functions can interact with objects of different types using the same interface, and the correct behavior is determined at runtime based on the object's actual class.


24 What is method chaining in Python OOP?
-Method chaining in Python Object-Oriented Programming (OOP) refers to the technique where multiple methods are called on the same object in a single line, one after another. This is possible because each method returns the object itself (or a modified version of it), allowing for further method calls.

In Python, this is often implemented by having each method return `self`, which allows for consecutive method calls on the same object. The advantage of method chaining is that it makes code more concise and can enhance readability, especially when multiple operations need to be performed on the same object.

### Example:

```python
class Calculator:
    def __init__(self, value=0):
        self.value = value

    def add(self, num):
        self.value += num
        return self

    def subtract(self, num):
        self.value -= num
        return self

    def multiply(self, num):
        self.value *= num
        return self

    def divide(self, num):
        if num != 0:
            self.value /= num
        return self

    def get_value(self):
        return self.value

# Using method chaining
calc = Calculator()
result = calc.add(5).subtract(2).multiply(3).divide(2).get_value()
print(result)  # Output: 4.5
```

### Breakdown:
- `add(5)` adds 5 to `self.value` and returns the `Calculator` object itself (`self`).
- `subtract(2)` subtracts 2 from `self.value` and returns the object again.
- `multiply(3)` multiplies `self.value` by 3 and returns the object.
- `divide(2)` divides `self.value` by 2 and returns the object.
- `get_value()` retrieves the final value.

By chaining these methods together, we can perform multiple operations in a single line, which makes the code cleaner and more expressive.



25 What is the purpose of the __call__ method in Python?
-The `__call__` method in Python is a special (or "magic") method that allows an instance of a class to be called as if it were a function. When you define a `__call__` method in a class, you can use instances of that class like functions. This can be particularly useful in scenarios where you want to give objects callable behavior or encapsulate function-like logic inside an object.

### Syntax
```python
class MyClass:
    def __call__(self, *args, **kwargs):
        # Code to be executed when the object is called
        pass
```

### How It Works
- When you create an instance of a class that defines `__call__`, you can call the instance as if it were a function.
- The `__call__` method will be triggered when you do so, and it can accept arguments (just like a regular function).

### Example

```python
class Adder:
    def __init__(self, start=0):
        self.start = start
        
    def __call__(self, x):
        return self.start + x

# Create an instance of Adder
add_five = Adder(5)

# Call the instance like a function
result = add_five(10)  # This triggers __call__(self, 10)
print(result)  # Output: 15
```

In this example:
- The `Adder` class has a `__call__` method that adds a number to `self.start`.
- The `add_five` object is created with a starting value of 5.
- When `add_five(10)` is executed, it calls `__call__` with the argument `10`, and it adds `5 + 10` to produce `15`.

### Use Cases for `__call__`
1. **Function Objects**: When you want an object to behave like a function. This is useful when you need the flexibility of a function but also want to maintain the state of an object.
2. **Decorators**: Objects can be used as decorators if they implement `__call__`, allowing them to be used with the `@decorator` syntax.

````python
                                #Practical Questions#


 1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog
that overrides the speak() method to print "Bark!".
# Parent class Animal
class Animal:
    def speak(self):
        print("This animal makes a sound.")

# Child class Dog that overrides speak()
class Dog(Animal):
    def speak(self):
        print("Bark!")

# Creating an instance of Animal and Dog
animal = Animal()
dog = Dog()

# Calling the speak method on both instances
animal.speak()  # Output: This animal makes a sound.
dog.speak()     # Output: Bark!




 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both
from abc import ABC, abstractmethod
import math

# Abstract class Shape
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# Derived class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Derived class Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Example usage
circle = Circle(5)  # Circle with radius 5
r




3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute
# Base class Vehicle
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type  # Attribute for the type of vehicle

    def display_type(self):
        print(f"This is a {self.type} vehicle.")

# Derived class Car that inherits from Vehicle
class Car(Vehicle):
    def __init__(self, vehicle_type, brand, model):
        super().__init__(vehicle_type)  # Call the parent class constructor to initialize 'type'
        self.brand = brand  # Attribute for the car's brand
        self.model = model  # Attribute for the car's model

    def display_car_info(self):
        print(f"This is a {self.brand} {self.model} car.")

# Further derived class ElectricCar that inherits from Car
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, model, battery_capacity):
        super().__init__(vehicle_type, brand, model)  # Call the parent class constructor to initialize 'type', 'brand', and 'model'
        self.battery_capacity = battery_capacity  # Attribute for battery capacity

    def display_battery_info(self):
        print(f"This electric car has a battery capacity of {self.battery_capacity} kWh.")

# Create an instance of ElectricCar
electric_car = ElectricCar("Electric", "Tesla", "Model S", 100)

# Display information about the vehicle
electric_car.display_type()  # Inherited from Vehicle
electric_car.display_car_info()  # Inherited from Car
electric_car.display_battery_info()  # Unique to ElectricCar




4. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute.
# Base class Vehicle
class Vehicle:
    def __init__(self, type_of_vehicle):
        self.type = type_of_vehicle  # Attribute to store vehicle type

    def display_info(self):
        return f"Vehicle Type: {self.type}"

# Derived class Car from Vehicle
class Car(Vehicle):
    def __init__(self, type_of_vehicle, model, color):
        super().__init__(type_of_vehicle)  # Initialize the base class
        self.model = model  # Additional attribute for Car
        self.color = color  # Additional attribute for Car

    def display_info(self):
        # Overriding the method to inc




 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance
class BankAccount:
    # Constructor to initialize the bank account with a balance
    def __init__(self, initial_balance=0.0):
        self.__balance = initial_balance  # Private attribute

    # Method to deposit money into the account
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount:.2f}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money from the account
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount:.2f}")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            print("Insufficient funds.")

    # Method to check the current balance
    def get_balance(self):
        return self.__balance

# Create an object of BankAccount class
account = BankAccount(1000.00)  # Initial balance of $1000

# Display the initial balance
print(f"Initial Balance: ${account




6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar
and Piano that implement their own version of play()
// Base class
class Instrument {
    // Base class play() method
    public void play() {
        System.out.println("Playing an instrument");
    }
}

// Derived class Guitar
class Guitar extends Instrument {
    // Overriding play() method for Guitar
    @Override
    public void play() {
        System.out.println("Playing the guitar");
    }
}

// Derived class Piano
class Piano extends Instrument {
    // Overriding play() method for Piano
    @Override
    public void play() {
        System.out.println("Playing the piano");
    }
}

public class Main {
    public static void main(String[] args) {
        // Creating objects of Guitar and Piano
        Instrument instrument1 = new Guitar();
        Instrument instrument2 = new Piano();

        // Demonstrating runtime polymorphism
        instrument1.play();  // Will call play() from Guitar
        instrument2.play();  // Will call play() from Piano
    }
}





7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers.
      class MathOperations:

    @classmethod
    def add_numbers(cls, num1, num2):
        """Class method to add two numbers."""
        return num1 + num2

    @staticmethod
    def subtract_numbers(num1, num2):
        """Static method to subtract two numbers."""
        return num1 - num2

# Example usage:
result_add = MathOperations.add_numbers(5, 3)  # Using class method to add numbers
result_subtract = MathOperations.subtract_numbers(5, 3)  # Using static method to subtract numbers

print("Sum:", result_add)         # Output: Sum: 8
print("Difference:", result_subtract)  # Output: Difference: 2



8. Implement a class Person with a class method to count the total number of persons created
      class Person:
    # Class variable to store the count of persons created
    total_persons = 0

    def __init__(self, name, age):
        self.name = name
        self.age = age
        # Increment the count every time a new instance is created
        Person.total_persons += 1

    # Class method to return the total number of persons
    @classmethod
    def get_total_persons(cls):
        return cls.total_persons

# Creating instances of Person
person1 = Person("Alice", 30)
person2 = Person("Bob", 25)
person3 = Person("Charlie", 35)

# Accessing the class method to get the total number of pers




9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the
fraction as "numerator/denominator".
      class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Example usage:
fraction = Fraction(3, 4)
print(fraction)  # Output: 3/4





 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors
      class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

    def __add__(self, other):
        # Adding the corresponding components of the two vectors
        return Vector(self.x + other.x, self.y + other.y)

# Creating two Vector objects
vector1 = Vector(2, 3)
vector2 = Vector(4, 5)

# Adding two vectors using overloaded + operator
result_vector = vector1 + vector2

# Printing the result
print(f"Vector 1: {vector1}")
print(f"Vector 2: {vector2}")
print(f"Result of addition: {result_vector}")





 11. Create a class Person with attributes name and age. Add a method greet() that prints "Hello, my name is
{name} and I am {age} years old.
      class Person:
    def __init__(self, name, age):
        self.name = name  # Attribute for name
        self.age = age    # Attribute for age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Example usage:
person1 = Person("Alice", 30)
person1.greet()  # Output: Hello, my name is Alice and I am 30 years old.





12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
the average of the grades.
      class Student:
    def __init__(self, name, grades):
        # Initialize the student's name and grades
        self.name = name
        self.grades = grades

    def average_grade(self):
        # Compute the average of the grades
        if len(self.grades) == 0:
            return 0  # Avoid division by zero if there are no grades
        return sum(self.grades) / len(self.grades)

# Example of usage:
student1 = Student("Alice", [85, 90, 78, 92])
print(f"Student Name: {student1.name}")
print(f"Average Grade: {student1.average_grade():.2f}")





13. Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the
area
      class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        """Set the dimensions of the rectangle."""
        self.length = length
        self.width = width

    def area(self):
        """Calculate the area of the rectangle."""
        return self.length * self.width

# Example usage:
rect = Rectangle()
rect.set_dimensions(5, 3)  # Set length to 5 and width to 3
print(f"Area of the rectangle: {rect.area()}")  # Output will be 15





14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary
      class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        # Salary = hours worked * hourly rate
        return self.hours_worked * self.hourly_rate


class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        # Call the constructor of the base class Employee
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        # Call the calculate_salary method from the Employee class and add bonus
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Example usage:
# Creating an Employee object
employee = Employee("John Doe", 160, 20)
print(f"Employee Salary: ${employee.calculate_salary()}")

# Creating a Manager object with an additional bonus
manager = Manager("Alice Smith", 160, 30, 2000)
print(f"Manager Salary: ${manager.calculate_salary()}")





15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that
calculates the total price of the product
      class Product:
    def __init__(self, name, price, quantity):
        self.name = name        # Product name
        self.price = price      # Price of one unit
        self.quantity = quantity  # Quantity of the product

    def total_price(self):
        # Calculate total price (price * quantity)
        return self.price * self.quantity

# Example usage
product = Product("Laptop", 1000, 3)
print(f"Total price for {product.name}: ${product.total_price()}")





16 Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method.
      from abc import ABC, abstractmethod

# Abstract class Animal
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Derived class Cow
class Cow(Animal):
    def sound(self):
        return "Moo"

# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        return "Baa"

# Example usage
cow = Cow()
sheep = Sheep()

print(f"Cow makes sound: {cow.sound()}")
print(f"Sheep makes sound: {sheep.sound()}")






 17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
returns a formatted string with the book's details.
class Book:
    def __init__(self, title, author, year_published):
        # Initialize the attributes
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        # Return a formatted string with book's details
        return f"Title: {self.title}\nAuthor: {self.author}\nYear Published: {self.year_published}"

# Example usage:
book1 = Book("1984", "George Orwell", 1949)
print(book1.get_book_info())






  18. Create a class House with attributes address and price. Create a derived class Mansion that adds an
attribute number_of_rooms
      # Base class House
class House:
    def __init__(self, address, price):
        self.address = address  # Address of the house
        self.price = price      # Price of the house

    def display_info(self):
        return f"House Address: {self.address}, Price: ${self.price}"

# Derived class Mansion
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        # Initialize the base class (House) with address and price
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms  # Number of rooms in the mansion

    def display_info(self):
        # Call the base class method to get address and price info
        house_info = super().display_info()
        return f"{house_info}, Number of Rooms: {self.number_of_rooms}"

# Example usage
mansion = Mansion("123 Luxury Lane", 5000000, 12)
print(mansion.display_info())










````
